```python
import json
import pandas as pd
```

```python
def create_data(path):
    data = []
    # Read the file line by line
    with open(path, 'r') as file:
        for line in file:
            # Parse each line as a JSON object
            patient_dict = json.loads(line.strip())

            # Extract id and class_name
            id = patient_dict['id'].split('_')[0]  # Remove the suffix
            class_name = patient_dict['class_name']

            # Extract entity_stats and flatten it
            entity_stats = patient_dict['entity_stats']
            flattened_stats = {key: value for key, value in entity_stats.items()}

            # Combine id, class_name, and flattened_stats into a single dictionary
            combined_data = {'id': id, 'class_name': class_name}
            combined_data.update(flattened_stats)

            # Append the combined data to the list
            data.append(combined_data)
    return data

```

```python
data = create_data('/content/train.txt')
df = pd.DataFrame(data)
df
```
